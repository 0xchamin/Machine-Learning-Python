There are two ways of creating MatPlotLib charts<br>
1. Functional Method <br>
2. Object Oriented Method

1. Functional Method

```python
plt.plot(x,y) 
#plt.plot()
#adding labl
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Sample Graph Using MatPlotLib')
```

Creating Multi-Plot in same canvas
<br>
subplot() method is used

```python
plt.subplot(1,2,1) #rows, columns #plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Object Oriented API method
<br>
The approach is to instantiate "figure" object.
<br>
And call methods and properties in that object

```python
fig = plt.figure() #instantiate a figure() object

#Pass the following as a list to the method
#length of left axis
#bottom of axis
#width 
#heights

#All the drawings added to the figure and axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X axis label')
axes.set_ylabel('Y axis label')
axes.set_title('Object Oriented Plot')
```

```python
fig = plt.figure()

#left bottom width and height 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

#left bottom width and height 
axes1 = fig.add_axes([0.1, 0.1,1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

In Object Oriented Approach <br>
1. Create a figure instance - Creat Canvas (empty canvas)<br>
2. On canvas create/add different size axes by passing values correspond to left, bottom, width, height
3. Plot on canvas axises
